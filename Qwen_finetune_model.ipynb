```python
# ✅ Download from official source
!wget http://www.cs.cornell.edu/~cristian/data/cornell_movie_dialogs_corpus.zip
!unzip -q cornell_movie_dialogs_corpus.zip

import os, json
from sklearn.model_selection import train_test_split

# ✅ Read movie lines and conversations
lines_path = "cornell movie-dialogs corpus/movie_lines.txt"
convos_path = "cornell movie-dialogs corpus/movie_conversations.txt"

id2line = {}
with open(lines_path, encoding="utf-8", errors="ignore") as f:
    for line in f:
        parts = line.strip().split(" +++$+++ ")
        if len(parts) == 5:
            id2line[parts[0]] = parts[4]

pairs = []
with open(convos_path, encoding="utf-8", errors="ignore") as f:
    for line in f:
        parts = line.strip().split(" +++$+++ ")
        if len(parts) == 4:
            utterance_ids = eval(parts[3])  # list of utterance ids
            for i in range(len(utterance_ids) - 1):
                a = id2line.get(utterance_ids[i], "")
                b = id2line.get(utterance_ids[i + 1], "")
                if a and b:
                    pairs.append({"prompt": a, "response": b})

# ✅ Split and save as JSONL
os.makedirs("Data", exist_ok=True)
train, val = train_test_split(pairs, test_size=0.1, random_state=42)

def save_jsonl(data, path):
    with open(path, "w", encoding="utf-8") as f:
        for item in data:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")

save_jsonl(train, "Data/train.jsonl")
save_jsonl(val, "Data/valid.jsonl")

print(f"✅ Dataset ready! Total: {len(pairs)} | Train: {len(train)} | Valid: {len(val)}")
```

```python
import json
from datasets import Dataset, DatasetDict

# ✅ Load train.jsonl manually
with open("Data/train.jsonl", "r", encoding="utf-8") as f:
    train_data = [json.loads(line) for line in f]

# ✅ Load valid.jsonl manually
with open("Data/valid.jsonl", "r", encoding="utf-8") as f:
    val_data = [json.loads(line) for line in f]

# ✅ Convert to HuggingFace Datasets
train_dataset = Dataset.from_list(train_data)
val_dataset = Dataset.from_list(val_data)

# ✅ Wrap into a DatasetDict
dataset = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset
})

# ✅ Sanity check
print(dataset)
print("Sample row:", dataset["train"][0])
```

```python
import json
import torch
import numpy as np
import evaluate
from transformers import AutoTokenizer, AutoModelForCausalLM
from datasets import Dataset

# ✅ Load model and tokenizer
model_id = "Qwen/Qwen-1_8B-Chat"
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", trust_remote_code=True)

# ✅ Set pad_token to <|endoftext|>
tokenizer.pad_token = "<|endoftext|>"
model.config.pad_token_id = tokenizer.pad_token_id

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.eval()

# ✅ Load validation data (local .jsonl file)
def read_valid_jsonl(path):
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            yield json.loads(line)

valid_data = list(read_valid_jsonl("/content/Data/valid.jsonl"))
dataset = Dataset.from_list(valid_data)

# ✅ Load ROUGE
rouge = evaluate.load("rouge")
print("🧪 Running evaluation on 3 samples...")
perplexities, rouge_scores = [], []

# ✅ Loop through 3 examples
for i, example in enumerate(dataset.select(range(3))):
    print(f"\n🔄 Sample {i+1}")
    prompt = example["prompt"]
    reference = example["response"]

    # Full input for perplexity
    full_input = f"<|user|>\n{prompt}\n<|assistant|>\n{reference}"
    inputs = tokenizer(full_input, return_tensors="pt", padding=True, truncation=True, max_length=512)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        loss = model(**inputs, labels=inputs["input_ids"]).loss.item()
    perplexities.append(np.exp(loss))

    # Prompt-only input for generation
    gen_prompt = f"<|user|>\n{prompt}\n<|assistant|>\n"
    gen_inputs = tokenizer(gen_prompt, return_tensors="pt", padding=True, truncation=True, max_length=512)
    gen_inputs = {k: v.to(device) for k, v in gen_inputs.items()}

    with torch.no_grad():
        output_ids = model.generate(**gen_inputs, max_new_tokens=50)
    reply = tokenizer.decode(output_ids[0], skip_special_tokens=True).split("<|assistant|>")[-1].strip()

    # ROUGE score
    scores = rouge.compute(predictions=[reply], references=[reference])
    rouge_scores.append(scores)

# ✅ Final Results
print("\n✅ Evaluation Complete!")
print(f"Avg Perplexity: {np.mean(perplexities):.2f}")
for k in rouge_scores[0]:
    print(f"{k}: {np.mean([s[k] for s in rouge_scores]):.4f}")
```

```python
# infer_qwen.py

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# ✅ Load Qwen model and tokenizer from Hugging Face
model_id = "Qwen/Qwen-1_8B-Chat"
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", trust_remote_code=True)

# ✅ Set pad_token to <|endoftext|> if not already set
tokenizer.pad_token = "<|endoftext|>"
model.config.pad_token_id = tokenizer.pad_token_id

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.eval()

# ✅ Interactive loop
print("🟢 Chatbot is ready! Type 'exit' to quit.\n")
while True:
    user_input = input("🧑 You: ")
    if user_input.strip().lower() in ["exit", "quit"]:
        print("👋 Goodbye!")
        break

    # Format conversation as per Qwen-style chat
    prompt = f"<|user|>\n{user_input}\n<|assistant|>\n"
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=100,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            pad_token_id=tokenizer.pad_token_id
        )

    # Decode and print the response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = response.split("<|assistant|>")[-1].strip()
    print("🤖 Bot:", response)
```

```python
# gradio_chatbot.py

import torch
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM

# ✅ Load Qwen model and tokenizer
model_id = "Qwen/Qwen-1_8B-Chat"
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", trust_remote_code=True)

# ✅ Set pad_token if needed
tokenizer.pad_token = "<|endoftext|>"
model.config.pad_token_id = tokenizer.pad_token_id

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.eval()

# ✅ Chat history support
def generate_reply(message, history):
    history = history or []
    prompt = ""
    for user, bot in history:
        prompt += f"<|user|>\n{user}\n<|assistant|>\n{bot}\n"
    prompt += f"<|user|>\n{message}\n<|assistant|>\n"

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            pad_token_id=tokenizer.pad_token_id
        )

    full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
    reply = full_output.split("<|assistant|>")[-1].strip()
    history.append((message, reply))
    return history, history

# ✅ Beautiful Gradio UI
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        <h1 style='text-align: center; color: #2F80ED;'>🤖 Qwen 1.8B Chatbot</h1>
        <p style='text-align: center;'>Ask anything! Powered by Qwen/Qwen-1_8B-Chat and Transformers.</p>
        """)

    chatbot = gr.Chatbot(height=400)
    msg = gr.Textbox(placeholder="Type your message here...", label="Your Message")
    clear = gr.Button("🧹 Clear Chat")

    state = gr.State([])

    msg.submit(generate_reply, [msg, state], [chatbot, state])
    clear.click(lambda: ([], []), None, [chatbot, state])

# ✅ Launch the Gradio app
demo.launch()
```
